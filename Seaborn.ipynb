### Categorical Plots

```python
sns.distplot(iris['sepal.length'],bins=30)
```

```python
sns.jointplot(x='sepal.length',y='sepal.width',data = iris,kind='hex')
```

```python
sns.pairplot(data=iris,hue='variety',palette='coolwarm')
```

```python
sns.rugplot(iris['sepal.length'])
```

### Matrix Plots

```python
ma = iris.corr()
ma
```

```python
sns.heatmap(data = ma, annot=True)
```

```python
data2 = pd.read_csv('SalesRecords.csv')
data2.head()
```

```python
d2 = data2.corr()
sns.heatmap(data = d2, annot = False)
```

```python
sls = data2.pivot_table(index='Item Type',columns='Region',values='Total Profit')
sls
```

```python
sns.heatmap(data = sls, linecolor='white',linewidths=0.5)
```

```python
sns.clustermap(data = sls,cmap='coolwarm')
```

### Grids
